# Linear Regression

## The Normal Equation

```python
import numpy as np

X = 2*np.random.rand(100, 1) 
y = 4 + 3*X + np.random.randn(100, 1)
```

```python
import matplotlib
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X] # add x0 = 1 to each instance; bias term 
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

Of course, ideally we should get b = 4, grad = 3 but the above isn't too shabby

```python
X_new = np.array([[0], [2]]) # predictions on these
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_pred = X_new_b.dot(theta_best)
y_pred
```

Plot prediction line (defined by two points) vs original data

```python
plt.plot(X_new, y_pred, "r-")
plt.plot(X, y, "b.")
plt.axis([0,2,0,15])
plt.show()
```

Linear Regression with sklearn

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
lin_reg.predict(X_new)
```

## Gradient Descent

```python
m = 100
X = 2*np.random.rand(m, 1)
y = 4 + 3*X + np.random.rand(m, 1)

X_b = np.c_[np.ones((m, 1)), X]
```

```python
eta = 0.1 # learning rate
n_iterations = 10000

theta_path_bgd = []

theta = np.random.randn(2, 1) # random initialization of bias and gradient
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
    theta_path_bgd.append(theta)
```

```python
theta
```

```python
# Illustrate that -= modifies arr in-place. 
# note the diff between theta -= .. and theta = theta - ..
x = np.array([1,2])
ret = []
ret.append(x)
x[0] = 5
ret
```

# Stochastic Gradient Descent

```python
n_epochs = 50
t0, t1 = 5, 50 # learning schedule hyperparams
theta_path_sgd = []


def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2, 1)

for epoch in range(n_epochs):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(m):
        xi = X_b_shuffled[i].reshape(1,2)
        yi = y_shuffled[i]
        gradients = 2*xi.T.dot(xi.dot(theta)-yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta*gradients
        theta_path_sgd.append(theta)
theta
```

```python
from sklearn.linear_model import SGDRegressor
# runs for 1k epochs or until loss drops by less than 0.001 during one epoch
sgd_reg = SGDRegressor(max_iter=10000, tol=1e-3, penalty=None, eta0=0.1) 
sgd_reg.fit(X, y.ravel())
```

## Mini-batch Gradient Descent

```python
n_epochs = 50
mini_batch_size = 20
t0, t1 = 200, 1000
theta = np.random.randn(2,1)
m = 100
theta_path_mgd = []

def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_epochs):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, mini_batch_size):
        t += 1
        xi = X_b_shuffled[i:i+mini_batch_size]
        yi = y_shuffled[i:i+mini_batch_size]
        gradients = 2/mini_batch_size * xi.T.dot(xi.dot(theta)-yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)
theta
```

```python
theta_path_bgd = np.array(theta_path_bgd)
theta_path_sgd = np.array(theta_path_sgd)
theta_path_mgd = np.array(theta_path_mgd)
```

```python
plt.figure(figsize=(7,4))
plt.plot(theta_path_sgd[:, 0], theta_path_sgd[:, 1], "r-s", linewidth=1, label="Stochastic")
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], "g-+", linewidth=2, label="Mini-batch")
plt.plot(theta_path_bgd[:, 0], theta_path_bgd[:, 1], "b-o", linewidth=3, label="Batch")
plt.legend(loc="upper left", fontsize=16)
plt.xlabel(r"$\theta_0$", fontsize=20)
plt.ylabel(r"$\theta_1$   ", fontsize=20, rotation=0)
plt.axis([3, 4.8, 2.3, 4])
plt.show()
```

## Polynomial Regression

```python
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.rand(m, 1)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
# Using closed form to solve polynomial regression
X_b = np.c_[np.ones((100,1)), X_poly]
np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
# data for plotting curve
X_show = np.arange(-3, 3.1, 0.1)
X_show = X_show.reshape(X_show.shape[0],1)
X_show_poly = poly_features.fit_transform(X_show)
y_pred = lin_reg.predict(X_show_poly)
```

```python
plt.plot(X_show, y_pred, "r-")
plt.plot(X, y, "b.")
plt.axis([-3,3,0,10])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

for style, width, degree in (("g-", 1, 300), ("b--", 2, 2), ("r-+", 2, 1)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([
            ("poly_features", polybig_features),
            ("std_scaler", std_scaler),
            ("lin_reg", lin_reg),
        ])
    polynomial_regression.fit(X, y)
    y_newbig = polynomial_regression.predict(X_show)
    plt.plot(X_show, y_newbig, style, label=str(degree), linewidth=width)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper left")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
plt.show()
```

## Ridge Regression (Regularized Linear Regression)

Regularization term: alpha/2 * square of its feature

**Note**: It is important to scale the data (eg. StandardScaler before performing Ridge Regression, as it is sensitive to the scale of the input features. This is true of most regularization techniques)

After all, regularization is only applied to the cost function during training. It should not be actively used after the model is trained and deployed for prediction.

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1, solver="cholesky")
ridge_reg.fit(X, y)
ridge_reg.predict([[1.5]])
```

```python
sgd_reg = SGDRegressor(penalty="l2") # ridge regression equivalent with gradient descent instead of closed form
sgd_reg.fit(X, y.ravel())
sgd_reg.predict([[1.5]])
```

## Lasso Regression

By differentiating, notice that lasso loss drops linearly. So Lasso tends to eliminate the weights of the least important features

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)
lasso_reg.predict([[1.5]])
```

```python
# Elastic Net is a middle ground between Lasso and Ridge

from sklearn.linear_model import ElasticNet

elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X, y)
elastic_net.predict([[1.5]])
```

## Early Stopping

```python
from sklearn.base import clone
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

m = 100
X = 4 * np.random.rand(m, 1) - 2
y = 0.7 * 2*X**5 + 3*X**4 + 0.5*X**2 + X + 2 + np.random.rand(m, 1)

X_train = X[:80]
y_train = y[:80]
X_val = X[80:]
y_val = y[80:]

# prepare data
poly_scaler = Pipeline([
    ("poly_features", PolynomialFeatures(degree=20, include_bias=False)),
    ("std_scaler", StandardScaler())
])

X_train_poly_scaled = poly_scaler.fit_transform(X_train)
X_val_poly_scaled = poly_scaler.fit_transform(X_val)

sgd_reg = SGDRegressor(max_iter=1, tol=None, warm_start=True, penalty=None, learning_rate="constant", eta0=0.0001)

min_val_error = float('inf')
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled, y_train.ravel()) # continues where it prev left off
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val, y_val_predict)
    if val_error < min_val_error:
        min_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)

best_epoch
```

## Logistic Regression

```python
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())
```

```python
X = iris["data"][:, 3:] # petal width
y = (iris["target"] == 2).astype(int) # 1 if iris virginica else 0
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X, y)
```

```python
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris Virginica")
```

```python
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]

plt.figure(figsize=(8, 3))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")
plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris virginica")
plt.text(decision_boundary+0.02, 0.15, "Decision  boundary", fontsize=14, color="k", ha="center")
plt.arrow(decision_boundary[0], 0.08, -0.3, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
plt.arrow(decision_boundary[0], 0.92, 0.3, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')
plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 3, -0.02, 1.02])
plt.show()
```

## Softmax Regression

```python
X = iris["data"][:, (2,3)]
y = iris["target"]

softmax_reg = LogisticRegression(multi_class="multinomial", solver="lbfgs", C=18)
softmax_reg.fit(X, y)
```

Below shows decision boundaries for _Iris versicolor_ class, represented by the curved lines

```python
x0, x1 = np.meshgrid(
        np.linspace(0, 8, 500).reshape(-1, 1),
        np.linspace(0, 3.5, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]


y_proba = softmax_reg.predict_proba(X_new)
y_predict = softmax_reg.predict(X_new)

zz1 = y_proba[:, 1].reshape(x0.shape) # show probabilities for class 1 which is iris versicolor
zz = y_predict.reshape(x0.shape) # show decision boundaries 

plt.figure(figsize=(10, 4))
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="Iris virginica")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="Iris versicolor")
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="Iris setosa")

from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])

plt.contourf(x0, x1, zz, cmap=custom_cmap)
contour = plt.contour(x0, x1, zz1, cmap=plt.cm.brg)
plt.clabel(contour, inline=1, fontsize=12)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 7, 0, 3.5])
plt.show()
```
